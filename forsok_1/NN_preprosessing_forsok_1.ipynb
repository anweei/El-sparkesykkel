```python
import numpy as np
import pandas as pd
```

## Restructuring raw dataframe ##

```python
df_acc = pd.read_csv('csv/raw_acc_data.csv')
df_gyro = pd.read_csv('csv/raw_gyro_data.csv')
```

```python
df_acc['label'] = df_acc['label'].map({'Normal': 0, 'Reckless': 1})
df_gyro['label'] = df_gyro['label'].map({'Normal': 0, 'Reckless': 1})
```

```python
df_acc
```

```python
df_gyro
```

## Accelerometer y direction

```python
acc_y_arr = df_acc['acc_y'].values
```

```python
acc_y_list = acc_y_arr.tolist()
```

```python
list_of_list = []
for i in range(948):
    temp = acc_y_list[i*400:(i+1)*400]
    list_of_list.append(temp)
```

```python
df_acc_y = pd.DataFrame(list_of_list) 
```

```python
df_acc_y
```

```python
labels = list(df_acc['label'].values)
y = []
for i in range(948):
    y.append(labels[i*400])
```

```python
df_acc_y['y']=y
```

```python
df_acc_y
```

```python
df_acc_y.to_csv('csv/df_acc_y.csv')
```
